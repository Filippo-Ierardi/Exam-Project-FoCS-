## 12) For each year, compute the total amount of loans. Each loan that has planned expiration time and disburse time in different years must have its amount distributed proportionally to the number of days in each year. For example, a loan with disburse time December 1st, 2016, planned expiration time January 30th 2018, and amount 5000USD has an amount of 5000USD * 31 / (31+365+30) = 363.85 for 2016, 5000USD * 365 / (31+365+30) = 4284.04 for 2017, and 5000USD * 30 / (31+365+30) = 352.11 for 2018.

L'idea è la seguente: per ogni prestito voglio ottenere un numero di righe pari al numero di anni che il prestito coinvolge, in seguito voglio calcolare il corrispondente ammontare del prestito per ogni anno (distribuito proporzionalmente al numero di giorni) e infine voglio raggruppare per anno e calcolare l'ammontare totale. 

Per prima cosa, estraggo gli anni di esborso e di scadenza dei prestiti dalle colonne disburse_time e planned_expiration_time con il metodo .dt.year. Creo quindi una copia del dataset loans prendendo solo le colonne di interesse ed escludendo le righe che presentano valori nulli (non è nota la data di esborso o la data di scadenza).

```python
loans['disburse_year'] = loans['disburse_time'].dt.year
```

```python
loans['expiration_year'] = loans['planned_expiration_time'].dt.year
```

```python
loans2 = loans[['loan_id','loan_amount', 'disburse_time', 'planned_expiration_time', 'disburse_year', 'expiration_year', 'duration']].dropna()
loans2
```

Voglio ottenere una colonna che ha come valori delle liste contenenti gli anni che il prestito coinvolge. Essendo disburse_year ed expiration_year dei float, converto le colonne in int in modo da poter usare la funzione nativa range().

```python
loans2['disburse_year'] = loans2['disburse_year'].astype(int)
loans2['expiration_year'] = loans2['expiration_year'].astype(int)
```

```python
loans2['year'] = [list(range(i, j+1)) for i, j in loans2[['disburse_year', 'expiration_year']].values]
loans2
```

Utilizzo il metodo .explode() sulla colonna year appena creata.

```python
loans2 = loans2.explode('year')
loans2
```

Per calcolare la porzione di un prestito corrispondente ad un particolare anno creo un'altra funzione sfruttando la libreria datetime. La funzione confronta le colonne year, disburse_year ed expiration_year e restituisce un valore diverso a seconda dei casi:
- Se disburse_year e expiration_year sono uguali, restistuisce l'intero ammontare del prestito
- Se disburse_year è maggiore di expiration_year, restituisce un valore nullo
- Se disburse_year è minore di expiration_year e year è uguale a disburse_year, calcola il rapporto tra il numero di giorni a partire da disburse_time fino alla fine dell'anno e la durata totale del prestito e moltiplica tale rapporto per loan_amount
- Se disburse_year è minore di expiration_year e year è uguale a expiration_year, calcola il rapporto tra il numero di giorni a partire dall'inizio dell'anno fino a planned_expiration_time e la durata totale del prestito e moltiplica tale rapporto per loan_amount
- Se disburse_year è minore di expiration_year e year è strettamente compreso tra disburse_year ed expiration_year, calcola il rapporto tra il numero di giorni dell'anno e la durata totale del prestito e moltiplica tale rapporto per loan_amount

```python
import datetime
```

```python
def amount_per_year(row):
    if row['disburse_year'] == row['expiration_year']:
        value = row['loan_amount']
    elif row['disburse_year'] > row['expiration_year']:
        value = None
    elif row['year'] == row['disburse_year']:
        value = round(row['loan_amount'] * (datetime.date(row['year'] + 1,1,1) - row['disburse_time'].date()).days / (row['duration'] + 1), 2)
    elif row['year'] == row['expiration_year']:
        value = round(row['loan_amount'] * (row['planned_expiration_time'].date() - datetime.date(row['year'] - 1, 12, 31)).days / (row['duration'] + 1), 2)
    elif row['year'] % 4:
        value = row['loan_amount'] * 365 / (row['duration'] + 1)
    else:
        value = row['loan_amount'] * 366 / (row['duration'] + 1)
    return value
```

Infine, applico la funzione creando la colonna loan_amount_per_year e raggruppo per anno sommando su tale colonna, ottenendo l'ammontare totale prestato per ogni anno.

```python
loans2['loan_amount_per_year'] = loans2.apply(amount_per_year, axis = 1)
loans2
```

```python
loans2[['year', 'loan_amount_per_year']].groupby('year').sum()
```
